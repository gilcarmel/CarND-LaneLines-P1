Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math
from scipy import stats

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    left_lane_lines = []
    right_lane_lines = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = ((y2-y1)/(x2-x1))
            if abs(slope) > 0.5:   #ignore mostly horizontal lines
                #cv2.line(img, (x1, y1), (x2, y2), color, thickness)
                # assign each segment to left or right lane according to its slope
                if slope < 0:
                    left_lane_lines.append(line)
                else:
                    right_lane_lines.append(line)

    #merge left/right lane segments into one line and draw them
    merge_and_draw_segments(img, left_lane_lines, color, thickness)
    merge_and_draw_segments(img, right_lane_lines, color, thickness)
             
# combines decected line segments into one solid lane line and draws it on img
def merge_and_draw_segments(img, segments, color, thickness):
    if len(segments) == 0:
        return
    
    # find best-fit line using linear regression
    x = []
    y = []
    for line in segments:
        for x1,y1,x2,y2 in line:
            x.append(x1)
            x.append(x2)
            y.append(y1)
            y.append(y2)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

    # draw the best-fit line from the bottom of the image to the middle
    height = img.shape[0]
    width = img.shape[1]
    y1 = height
    x1 = int(float(y1 - intercept) / slope)
    y2 = int(height * 0.6)
    x2 = int(float(y2 - intercept) / slope)
    cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
```

## Test on Images

Now you should build your pipeline to work on the images in the directory "test_images"  
**You should make sure your pipeline works well on these images before you try the videos.**

```python
import os

def process_image(path, output_path):
    #Read image
    img = mpimg.imread(path)
    plt.figure()
    plt.imshow(img)
    print('This image is:', type(img), 'with dimesions:', img.shape)

    # Grayscale + Guassian blur
    blurred_img = gaussian_blur(grayscale(img), 5)
    plt.figure()
    plt.imshow(blurred_img,  cmap='gray')

    # Edge detection
    canny_img = canny(blurred_img, 50, 150)
    plt.figure()
    plt.imshow(canny_img,  cmap='gray')
    
    # Apply region-of-interest mask
    height = img.shape[0]
    width = img.shape[1]
    mask_verts = np.array( [[[width*0.1,height],[width*0.4,height*0.6],[width*0.6,height*0.6],[width*0.9,height]]], dtype=np.int32 )
    masked_image = region_of_interest(canny_img, mask_verts)
    plt.figure()
    plt.imshow(masked_image)

    #Hough line detection
    rho = 2
    theta = math.pi / 180
    threshold = 80
    min_line_len = 40
    max_line_gap = 130
    lines_img = hough_lines(masked_image, rho, theta, threshold, min_line_len, max_line_gap)
    plt.figure()
    plt.imshow(lines_img)
    output_img = weighted_img(lines_img, img, =1., =1.)
    plt.figure()
    plt.imshow(output_img)
    
    mpimg.imsave(output_path, output_img)
    
#Process each input image and save the results under test_images/output/
input_path = "test_images/"
output_path = os.path.join(input_path, "output")
if not os.path.exists(output_path):
    os.mkdir(output_path)
for image_filename in os.listdir(input_path):
    input_imgname = os.path.join(input_path, image_filename)
    if not os.path.isdir(input_imgname):      #ignore output directory
        try:
            output_imgname = os.path.join(output_path, image_filename)
            process_image(input_imgname, output_imgname)
        except OSError:
            pass  #ignore invalid files
```

The current algorithm makes several assumptions that are likely to cause it to fail under less ideal circumstances:
* The shape mask assumes we're already in the middle of a lane. I don't think it would work well while the car is changing lanes.
* The shape mask also assumes a relatively level road (i.e. horizon is always at about 0.6 of image height). This wouldn't work when going downhill or uphill.
* The lane-line extrusion assumes one linear line will capture all detected line segments for each side of the lane. This would not work will in a curve. 
* The parameters for edge- and line-detection are hard-coded to work with the very limited set of daytime images. They would not work as well in other lighting conditions.

To help with detection when not centered in a lane, we could examine all detected lane lines and pick the two that are closest to horizontal as our current/target lane.

To help with curved roads, we could try to fit a curve to the detected line segments instead of a straight line.

To help with uphill/downhill roads, we might try to detect the horizon and use it to form our shape mask (rather than hard-coding it)

To help the pipeline perform better in other lighting conditions, we might run it with several sets of canny/hough paramemters, and add some heuristic for determining which set of parameters performed best on a given frame.
